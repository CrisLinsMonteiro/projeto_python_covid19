```python
import requests as r
```

```python
import csv
```

```python
import datetime as dt
```

```python
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
```

```python
raw_data[0]
```

```python
final_date = []
for obs in raw_data:
    final_date.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])
```

```python
final_date.insert(0, ['confirmados', 'obitos', 'recuperados', 'ativos', 'data'])
final_date
```

```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4
```

```python
for i in range(1, len(final_date)):
    final_date[i][DATA] = final_date[i][DATA][:10]
```

```python
final_date
```

```python
with open('brasil-covid.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(final_date)
```

```python
for i in range(1, len(final_date)):
    final_date[i][DATA] = dt.datetime.strptime(final_date[i][DATA], '%Y-%m-%d')
```

```python
final_date
```

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
          return [
            {
                'label': labels[0],
                'data': y
            }
         ]
```

```python
def set_title(title=''):
  if title != '':
      display = 'true'
  else:
      display = 'false'
  return {
      'title': title,
      'display': display
  }
```

```python
def create_chart(x, y, labels, kind='bar', title=''):

  datasets = get_datasets(y, labels)
  options = set_title(title)

  chart = {
      'type': kind,
      'data': {
          'labels': x,
          'datasets': datasets
      },
      'options': options
  }
  return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for obs in final_date[1::30]:
    y_data_1.append(obs[CONFIRMADOS])

y_data_2 = []
for obs in final_date[1::30]:
    y_data_2.append(obs[RECUPERADOS])   

labels = ['Confirmados', 'Recuperados']

x = []
for obs in final_date[1::10]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))

chart = create_chart(x, [y_data_1, y_data_2], labels, title='Grfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('/content/'+'meu-primeiro-grafico.png')
```

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
    text = quote(link) # parsing do link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
